```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, log_loss

from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df_train=pd.read_csv("./dataset/train.csv")
df_test=pd.read_csv("./dataset/test.csv")
print(df_train.shape,df_test.shape)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
le=LabelEncoder()
labels = le.fit_transform(df_train.species)
classes = list(le.classes_) 
```

```python
uniquelabels= np.unique(labels)
uniquelabels
```

```python
classes
```

```python
X=df_train.drop(['id', 'species'], axis=1).values
Y=labels
```

```python
labels.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20,shuffle=True,stratify=labels)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.pipeline import make_pipeline
clf1 =  GaussianNB()
clf1.fit(x_train, y_train)
y_pred1=clf1.predict(x_test)
pp1=clf1.predict_proba(x_test)
ll1= log_loss(y_test, pp1)

print('Accuracy: {:.4%}'.format(accuracy_score(y_test, y_pred1)))
print('Log Loss: {:.6}'.format(ll1))
```

```python
clf2=make_pipeline(StandardScaler(), LogisticRegression(solver='newton-cg', multi_class='multinomial'))
clf2.fit(x_train, y_train)
y_pred2=clf2.predict(x_test)

pp2=clf2.predict_proba(x_test)
ll2= log_loss(y_test, pp2)

print('Accuracy: {:.4%}'.format(accuracy_score(y_test, y_pred2)))
print('Log Loss: {:.6}'.format(ll2))
```

```python
clf3=make_pipeline(StandardScaler(), KNeighborsClassifier(3))
clf3.fit(x_train, y_train)
y_pred3=clf3.predict(x_test)
pp3=clf3.predict_proba(x_test)
ll3 = log_loss(y_test, pp3)

print('Accuracy: {:.4%}'.format(accuracy_score(y_test, y_pred3)))
print('Log Loss: {:.6}'.format(ll3))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf4= DecisionTreeClassifier(criterion='entropy',random_state=2)
# clf4=make_pipeline(StandardScaler(), DecisionTreeClassifier(3))
clf4.fit(x_train, y_train)
y_pred4=clf4.predict(x_test)
pp4=clf4.predict_proba(x_test)
ll4 = log_loss(y_test, pp4)

print('Accuracy: {:.4%}'.format(accuracy_score(y_test, y_pred4)))
print('Log Loss: {:.6}'.format(ll4))
```
